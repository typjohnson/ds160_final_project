```python
# importing all libraries that may be used
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Saving the dataset to a variable and checking it out

```python
df = pd.read_csv('eq2015.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
print(f'There are {df.shape[0]} rows and {df.shape[1]} columns currently in the dataset')
```

```python
#printing out column names
df.columns
```

```python
#dropping a few columns that will not be relevant to analysis
df = df.drop(columns=['building_id', 'vdcmun_id', 'ward_id'])
```

```python
#checking data types for columns
df.info()
```

```python
#creating two new columns that show the height and floors lost from earthquake
df['eq_height_difference'] = df['height_ft_pre_eq'] - df['height_ft_post_eq']
df['eq_floor_difference'] = df['count_floors_pre_eq'] - df['count_floors_post_eq']
```

```python
#removing errorneous rows from the data set 
neg_removal = df[(df['eq_height_difference'] < 0) | (df['eq_floor_difference'] < 0)].index
df.drop(neg_removal, inplace = True)
```

```python
df.shape
```

```python
df.sample(10)
```

```python
#for loop checks the amount of missing values in each column
for column_name in df.columns:
    missing_count = df[column_name].isna().sum()
    if (missing_count > 0):
        print(f"In '{column_name}' there are {missing_count} missing values")
```

```python
#imputing missing values with the mode (all are categorical)
#for loop iterates through each column and fill if missing
for column_name in df.columns:
    column_mode = df[column_name].mode()[0]
    df[column_name].fillna(column_mode, inplace=True)
#double checking to make sure there are no remaining missing values in each column 
for column_name in df.columns:
    missing_count = df[column_name].isna().sum()
    if (missing_count > 0):
        print(f"In '{column_name}' there are {missing_count} missing values")
    else: 
        print(f"No missing values in '{column_name}'")
```

# Exploratory Analysis

```python
#creating a bar chart to visualize the number of buildings in each district 
district_counts = df['district_id'].value_counts()
plt.figure(figsize=(10,4))  
district_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Buildings per District')
plt.xlabel('District ID')
plt.ylabel('Number of Buildings')
plt.xticks(rotation = 45)
plt.show()
```

```python
building_avg_age = df.groupby('district_id')['age_building'].mean()
plt.figure(figsize=(,4))  
building_avg_age.plot(kind='bar', color='red', edgecolor='black')
plt.title('Average Age of Building by District')
plt.xlabel('District ID')
plt.ylabel('Average Age of Building')
plt.xticks(rotation = 45)
plt.show()
```
